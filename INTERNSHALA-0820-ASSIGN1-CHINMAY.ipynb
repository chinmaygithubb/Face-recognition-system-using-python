```python
from skimage.feature import hog
from skimage import data, exposure
import face_recognition
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.patches as Rectangle
import matplotlib.patches as Circle
import numpy as np
import cv2 
import face_recognition
%matplotlib inline
```

```python
image = cv2.imread('George.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
face_locations = face_recognition.face_locations(image)
number_of_faces = len(face_locations)
print("Found {} face(s) in input image".format(number_of_faces))
```

```python
plt.imshow(image)
ax = plt.gca()
for face_location in face_locations:
    top, right,bottom, left = face_location
    x,y,w,h = left, top, right, bottom
    print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(x,y,w,h))
    rect = patches.Rectangle((x,y), w-x, h-y, fill = False, color = 'red')
    ax.add_patch(rect)
plt.show()
```

```python

image = cv2.imread('George.jpg')
face_demo = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image1 = cv2.imread('kajal.jpg')
kajal = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
# plt.imshow(image1)

image2 = cv2.imread('shruti.jpg')
shruti = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
# plt.imshow(image2)
```

```python
face_demo_encoding = face_recognition.face_encodings(face_demo)[0]
kajal_encoding = face_recognition.face_encodings(kajal)[0]
shruti_encoding = face_recognition.face_encodings(shruti)[0]

known_face_encodings = [
    face_demo_encoding,
    kajal_encoding,
    shruti_encoding
]
```

```python
image = cv2.imread('shruti1.jpg')
unknown_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(unknown_image)
unknown_face_encodings = face_recognition.face_encodings(unknown_image)
```

```python
from scipy.spatial import distance
for unknown_face_encoding in unknown_face_encodings:
    results = []
    for known_face_encoding in known_face_encodings:
        d = distance.euclidean(known_face_encoding, unknown_face_encoding)
        results.append(d)
    threshold = 0.6
    results = np.array(results) <= threshold
    
    name = 'Unknown'
    
    if results[0]:
        name = 'Face demo image'
    elif results[1]:
        name = 'shruti'
    elif results[2]:
        name = 'kajal'
    print('Found {} in this photo'.format(name))     
```
